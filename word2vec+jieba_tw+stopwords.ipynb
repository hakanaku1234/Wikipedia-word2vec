## 使用jieba_tw斷詞及去除停用詞

```python
import sys
sys.path.append('jieba_tw')
import jieba_tw
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# set jieba dict
jieba_tw.set_dictionary('jieba_dict/dict.txt.big')

# set stopwords
stopword_set = set()
with open('jieba_dict/stopwords.txt','r', encoding='utf-8') as stopwords:
    for stopword in stopwords:
        stopword_set.add(stopword.strip('\n'))

texts_num = 0

with open('jiebaTW_wiki_article/wiki_seg.txt', 'w', encoding='utf8') as output:
    with open('jiebaTW_wiki_article/zh_wiki_text.txt','r', encoding='utf8') as content :
        for line in content:
            words = jieba_tw.cut(line, cut_all=False)
        
            for word in words:
                if word not in stopword_set:
                    output.write(word + ' ')
            output.write('\n')
                
            texts_num += 1
        
            if texts_num % 10000 == 0:
                logging.info("已完成前 %d 行的斷詞" % texts_num)
```

### Word2vec

```python
from gensim.models import word2vec
import logging
import multiprocessing

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

sentences = word2vec.LineSentence("jiebaTW_wiki_article/wiki_seg.txt")

# sg=0, CBOW算法(大樣本)
# negative=5, 採用negativesampling, 有5個noise words(小樣本：5-20, 大樣本：2-5)
model = word2vec.Word2Vec(sentences, size=250, window=5, min_count=5,
                          workers=multiprocessing.cpu_count(), sg=0, negative=5)
```

```python
model.save("model/20180309wiki_model.bin")
```
